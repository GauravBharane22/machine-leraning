```python
#import
import pandas as pd
import numpy as np
```

```python
housing=pd.read_csv('Housing.csv')
```

```python
housing.head(20)
```

```python
housing.shape
```

```python
housing.info()
```

```python
housing.isnull().sum()
```

```python
# label encoding
varlist=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']

def convert(x):
    return x.map({'yes':1,'no':0}) #map is used to apply the change on whole data

housing[varlist]=housing[varlist].apply(convert) # apply the change , by using some function
```

```python
housing.head()
```

```python
status=pd.get_dummies(housing['furnishingstatus'])
```

```python
status.head()
```

```python
housing=pd.concat([housing,status],axis=1)
```

```python
housing.head()
```

```python
housing.drop(['furnishingstatus'],axis=1,inplace=True)
```

```python
housing.head()
```

```python
# Extract the dependent & independent var
X=housing.iloc[:,1:14]
```

```python
X
```

```python
# Extract the dependent & independent var
y=housing.iloc[:,0]
y
```

```python
# split the dataset into training & testing set
# machine learning
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=32)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Apply linear regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)# training to object on training data
```

```python
#y_test which actual output of x_test
y_pred=regressor.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test,y_pred))
```

```python
from sklearn import metrics
print(np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```python
# test score r2 score
from sklearn import metrics
print(metrics.r2_score(y_test,y_pred)) # 71% model is accurate 
```
